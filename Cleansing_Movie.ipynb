#impute gross, budget, dan num_critic_for_reviews dengan median

```python
#cek nilai median
movie["gross"].median()
```

```python
movie["title_year"].mode()[0]
```

```python
# impute gross, budget, dan num_critic_for_reviews dengan median
movie["gross"] = movie["gross"].fillna(movie["gross"].median())
movie["budget"] = movie["budget"].fillna(movie["budget"].median())
movie["num_critic_for_reviews"] = movie["num_critic_for_reviews"].fillna(movie["num_critic_for_reviews"].median())

# impute title_year dengan modus
movie["title_year"] = movie["title_year"].fillna(movie["title_year"].mode()[0])
```

```python
#melihat data uniq pada kolom num_user_for_reviews
movie['num_user_for_reviews'].unique()
```

```python
#mengganti nilai nan dan string kosong dg nilai 0
movie["num_user_for_reviews"] = movie["num_user_for_reviews"].fillna(0)
movie["num_user_for_reviews"] = movie["num_user_for_reviews"].replace({" ": 0})
```

```python
movie['num_user_for_reviews']
```

```python
# casting tipe data
movie["num_user_for_reviews"] = movie["num_user_for_reviews"].astype("int64")
movie["gross"] = movie["gross"].astype("int64")
movie["budget"] = movie["budget"].astype("int64")
movie["num_critic_for_reviews"] = movie["num_critic_for_reviews"].astype("int64")
movie["title_year"] = movie["title_year"].astype("int64")
movie.info()
```

```python
sns.histplot(data=movie, x="num_user_for_reviews")
```

```python
#mengganti nilai 0 dg nilai median dari data
movie["num_user_for_reviews"] = movie["num_user_for_reviews"].replace({0: movie["num_user_for_reviews"].median()})
```
